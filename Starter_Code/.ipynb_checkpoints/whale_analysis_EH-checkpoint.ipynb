```python
# Imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
whale_returns_df = pd.read_csv(whale_returns_csv, index_col='Date', parse_dates=True, infer_datetime_format=True) 
```

```python
# Count nulls
whale_returns_df.isna().sum()
```

```python
# Drop nulls
whale_returns_df = whale_returns_df.dropna()
whale_returns_df = whale_returns_df.sort_values('Date',ascending=True)
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
```

```python
# Count nulls
algo_returns_df.isna().sum()
```

```python
# Drop nulls
algo_returns_df = algo_returns_df.dropna()
algo_returns_df = algo_returns_df.sort_values('Date',ascending=True)
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_price_df = pd.read_csv(sp500_history_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
sp500_price_df = sp500_price_df.sort_values('Date',ascending=True)
sp500_price_df.head()
```

```python
# Check Data Types
sp500_price_df.dtypes
```

```python
# Fix Data Types
sp500_price_df['Close'] = sp500_price_df['Close'].str.replace('$','')
sp500_price_df['Close'] = sp500_price_df['Close'].astype('float64')
sp500_price_df.dtypes
```

```python
# Calculate Daily Returns
sp500_returns_df = sp500_price_df.pct_change()
sp500_returns_df.head()
```

```python
# Drop nulls
sp500_returns_df = sp500_returns_df.dropna()
sp500_returns_df.head()
```

```python
# Rename Column
sp500_returns_df.columns = ['S&P 500']
sp500_returns_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
all_dataframes = [whale_returns_df,algo_returns_df,sp500_returns_df]
combined_returns_df = pd.concat(all_dataframes,axis='columns', join='inner')
combined_returns_df.head()
```

# Portfolio Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500? 

```python
# Plot daily returns
combined_returns_df.plot(figsize=(15,7))
```

```python
# Plot cumulative returns
cumulative_returns_df = (1 + combined_returns_df).cumprod()
cumulative_returns_df.plot(figsize=(15,7))
```

#### Best Returning Portfolio

The best performing portfolio is Algo 1. However, both Algo 1 and Berkshire Hathaway Inc beat the S&P 500 over this time period.

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
combined_returns_df.boxplot(rot=45,figsize=(15,7))
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
combined_returns_df.std().sort_values(ascending=False)
```

```python
# Determine which portfolios are riskier than the S&P 500
combined_returns_df.std() > combined_returns_df['S&P 500'].std()
```

```python
# Calculate the annualized standard deviation (252 trading days)
combined_returns_df.std()*np.sqrt(252)
```

### Risk Metrics Analysis

Boxplot: The largest spread appears to be Berkshire Hathaway Inc. returns, while the lowest spread is in the Algo 1 returns.

Standard Deviations: Only Tiger Global Management and Berkshire Hathaway Inc have higher standard deviations than the S&P 500.

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
combined_returns_df.rolling(window=21).std().plot(figsize=(15,7))
```

```python
# Construct a correlation table
combined_returns_df.corr()
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
rolling_covariance_algo1 = combined_returns_df['Algo 1'].rolling(window=60).cov(combined_returns_df['S&P 500'])
rolling_variance_sp500 = combined_returns_df['S&P 500'].rolling(window=60).var()
rolling_beta_algo1 = rolling_covariance_algo1/rolling_variance_sp500
rolling_beta_algo1.plot(figsize=(15,7))
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
combined_returns_df.ewm(halflife=21).mean().plot(figsize=(15,10))
```

## Sharpe Ratios
In reality, investment managers and their institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (combined_returns_df.mean()*252) / (combined_returns_df.std()*np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind='bar')
```

On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales? Type your answer here:

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
amzn_csv = Path("Resources/amzn_historical.csv")
amzn_df = pd.read_csv(amzn_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
amzn_df.columns = ['AMZN']
amzn_df.head()
```

```python
# Read the second stock
goog_csv = Path("Resources/goog_historical.csv")
goog_df = pd.read_csv(goog_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
goog_df.columns = ['GOOG']
goog_df.head()
```

```python
# Read the third stock
aapl_csv = Path("Resources/aapl_historical.csv")
aapl_df = pd.read_csv(aapl_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
aapl_df.columns = ['AAPL']
aapl_df.head()
```

```python
# Read the fourth stock
dis_csv = Path("Resources/dis_historical.csv")
dis_df = pd.read_csv(dis_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
dis_df.columns = ['DIS']
dis_df.head()
```

```python
# Read the fifth stock
cost_csv = Path("Resources/cost_historical.csv")
cost_df = pd.read_csv(cost_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
cost_df.columns = ['COST']
cost_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
my_stocks = [amzn_df,aapl_df,goog_df,dis_df,cost_df]
my_stocks_df = pd.concat(my_stocks,axis='columns',join='inner')
my_stocks_df.index = pd.to_datetime(my_stocks_df.index, format = '%Y-%m-%d').strftime('%Y-%m-%d')
my_stocks_df.head()
```

```python
# Reset the index
my_stocks_df = my_stocks_df.reset_index()
my_stocks_df.head()
```

```python
# Pivot so that each column of prices represents a unique symbol
my_stocks_df = my_stocks_df.pivot_table(index="Date")
my_stocks_df = my_stocks_df.pct_change()
my_stocks_df.head()
```

```python
# Drop Nulls
my_stocks_df.dropna(inplace=True)
my_stocks_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/5, 1/5, 1/5, 1/5, 1/5]
my_portfolio_returns = my_stocks_df.dot(weights)
my_portfolio_returns = my_portfolio_returns.to_frame()
my_portfolio_returns.columns = ['Custom']
my_portfolio_returns.tail()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
combined_returns_df.index = pd.to_datetime(combined_returns_df.index, format = '%Y-%m-%d').strftime('%Y-%m-%d')
new_returns_df = pd.concat([combined_returns_df,my_portfolio_returns],axis='columns',join='inner')
new_returns_df.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
new_returns_df.dropna(inplace=True)
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Plot daily returns
new_returns_df.plot(figsize=(15,7))
```

```python
# Plot cumulative returns
new_cumulative_returns_df = (1 + new_returns_df).cumprod()
new_cumulative_returns_df.plot(figsize=(15,7))
```

```python
# Risk
# Box plot to visually show risk
new_returns_df.boxplot(rot=45,figsize=(15,7))
```

```python
# Daily Standard Deviations
new_returns_df.std().sort_values(ascending=False)
```

```python
# Determine which portfolios are riskier than the S&P 500
new_returns_df.std() > new_returns_df['S&P 500'].std()
```

```python
# Calculate the annualized standard deviation (252 trading days)
new_returns_df.std()*np.sqrt(252)
```

```python
# Rolling
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
new_returns_df.rolling(window=21).std().plot(figsize=(15,7))
```

```python
# Construct a correlation table
new_returns_df.corr()
```

```python
# Calculate Beta for two portfolios compared to the total market (S&P 500)
# Calculate Beta for custom portfolio
rolling_covariance_custom = new_returns_df['Custom'].rolling(window=60).cov(new_returns_df['S&P 500'])
rolling_variance_sp500 = new_returns_df['S&P 500'].rolling(window=60).var()
rolling_beta_custom = rolling_covariance_custom/rolling_variance_sp500

# Calculated Beta for algo 1
rolling_covariance_algo1 = new_returns_df['Algo 1'].rolling(window=60).cov(new_returns_df['S&P 500'])
rolling_beta_algo1 = rolling_covariance_algo1/rolling_variance_sp500

# Plot both Betas over time
initial_plot = rolling_beta_custom.plot(figsize=(15,7))
rolling_beta_algo1.plot(ax=initial_plot,figsize=(15,7), title="Beta Comparison to S&P 500")
initial_plot.legend(['Custom','Algo 1'])
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
new_returns_df.ewm(halflife=21).mean().plot(figsize=(15,10))
```

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (new_returns_df.mean()*252) / (new_returns_df.std()*np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind='bar')
```
